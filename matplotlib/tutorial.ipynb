```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 5, 0.1)
y = np.sin(x)
plt.plot(x, y)
# khởi tạo func figure trong file pyplot, có FigureClass đẫn đến file figure
# sau đó gọi func axes, kết thừa class trong file _axes.py
```

Các hàm luôn tạo ra figure và axes
- figure là khung
    - .subtitle() -> figure.py
- axes là trục tọa độ để vẽ bên trong
    - .set() -> matplotlib.axes.Axes.set

Các hàm tạo biểu đồ sẽ dùng:
- plt.subplots()

### **.subplots()**

```python
#figsize: size hình ảnh
x = np.linspace(0, 10, 100)
y = 4 + 2 * np.sin(2 * x)

fig, axs = plt.subplots(figsize=(4,3)) 

axs.plot(y, '--') # dùng func plot trong file _axes.py
axs.set(xlabel='range', ylabel='distance', title="image")
axs.set_title("title2")

```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y = 4 + 2 * np.sin(2 * x)

fig, axs = plt.subplots(title='four',xlabel = 'x') # def subplots trong file pyplot tạo ra figure và axs tại def figure

fig.suptitle('Figure')
axs.plot(y, 'r+') # dùng func plot trong file _axes.py
# xung đột 'r+' (bao gồm fmt = '[marker][line][color]'), marker='H', sẽ ưu tiên arg marker
axs.plot(x, 'r+', marker='H',color='green',linestyle = 'dashed', linewidth=1, markersize=6)
axs.set(xlim=(0, 8), xticks=np.arange(1, 20),
       ylim=(2, 15), yticks=np.arange(1, 10))
```
