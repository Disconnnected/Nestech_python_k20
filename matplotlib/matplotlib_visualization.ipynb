```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Global API

```python
x = np.arange(-10, 11)
```

```python
plt.figure(figsize=(16,9))
plt.title('graph function')

# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html
# marker|line|color

# plt.plot(x, x ** 2)
# plt.plot(x, x ** -2)
# plt.plot(x, -1 * (x ** 3 + x**2 -13))
plt.plot(x, -1 * (x ** 3 + x**2 -13), "or")
# plt.plot(x, -1 * (x ** 3 + x**2 -13), "")
```

```python
#marker
plt.figure(figsize=(16,9))
plt.title('about marker')


plt.plot(x, x ** 2, "o")

# markersize = ms = int
# markeredgecolor  = mec = str
# markerfacecolor = mfc = str

plt.plot(x, x ** 2, marker = "o", ms = 10, mec = "#4CAF50", c = "yellow")
```

```python
# line
plt.figure(figsize=(12,6))
plt.title('Line')

plt.plot(x, x ** 2, linestyle = 'dotted')
plt.plot(x, x ** 2, linestyle = 'dashed', c = "yellow")
plt.plot(x, x ** 2, linestyle = '')
plt.plot(x, x ** 2, linestyle = '-', linewidth = 10)
```

```python
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}
# plt.title("Graph of function level 2", fontdict = font1)
plt.title("Graph of function level 2", fontdict = font1, loc="right")
plt.xlabel("X value", fontdict = font2)
plt.ylabel("X ** 2 value", fontdict = font2)
plt.grid(axis = "y")
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
plt.plot(x, x ** 2)
```

```python
# subplot()
# With the subplot() function you can draw multiple plots in one figure:
```

```python
plt.figure(figsize=(12,6))

# plot #1
x1 = np.arange(-10,11)
# y1 = np.sin(x1)
y1 = x1**2
plt.subplot(1, 2, 1)
plt.title("Power of x")
plt.plot(x1,y1)
#the figure has 1 row, 2 columns, and this plot is the first plot.

# plot #2
# y2 = np.cos(x2)
y2 = np.arange(-10,11)
x2 = y2 ** 2
plt.subplot(1, 2, 2)
plt.title("Power of y")
plt.plot(x2,y2)
#the figure has 1 row, 2 columns, and this plot is the second plot.


plt.suptitle("Draw 2 subfigure")
```

### Scatter
The scatter() function plots one dot for each observation. It needs two arrays of the same length, one for the values of the x-axis, and one for values on the y-axis:

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.plot(x, y)
```
