```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
noi_lam = pd.read_csv('khaosat_noilam.csv')
met_moi = pd.read_csv('khaosat_metmoi.csv')
thoi_gian = pd.read_csv('khaosat_thoigian.csv')
viec_lam = pd.read_csv('khaosat_vieclam.csv')
```

```python
gr_noi_lam = noi_lam.groupby('answer').size()
print(gr_noi_lam.plot.pie(autopct='%1.1f%%',startangle=90, explode = [0,0.2,0,0,0],shadow = True))
plt.suptitle('Anh/chị đang làm ở đâu?')
plt.show()
```

```python
gr_met_moi = met_moi.groupby('answer').size()
print(gr_met_moi.plot.pie(autopct='%1.1f%%',startangle=90, explode = [0,0.1,0],shadow = True))
plt.suptitle('Hàng tuần anh/chị có cảm thấy đau đầu mỗi khi xếp lịch làm cho nhân viên không?')
plt.show()
```

```python
gr_thoi_gian = thoi_gian.groupby('answer').size()
print(gr_thoi_gian.plot.pie(autopct='%1.1f%%',startangle=90, explode = [0,0.1,0],shadow = True))
plt.suptitle('Anh/Chị có mất nhiều thời gian cho việc này không?')
plt.show()
```

```python
gr_viec_lam = viec_lam.groupby('answer').size()
print(gr_viec_lam.plot.pie(autopct='%1.1f%%',startangle=90))
plt.suptitle('Khi xếp lịch làm, anh/chị phải?')
plt.show()
```
